```python
wb_data = []
country_name = []
country_id = []
country_region = []
country_inc_level_id = []
country_inc_level_desc = []

wb_response = requests.get(url)
#print(wb_response.url)
wb_json = wb_response.json()
#l = len(wb_json["pages"])
#print(l)
#print(json.dumps(wb_json, indent=4))
for i in range(len(wb_json)):
    for item in wb_json[1]:
        wb_data.append({"Name":item['name'],
                       "Id":item['id'],
                       "Region":item['adminregion']['id'],
                       "IncomeLevelID":item['incomeLevel']['id'],
                       "IncomeLevelDesc":item['incomeLevel']['value']})
        
wb_data_df = pd.DataFrame.from_dict(wb_data)
#Export to csv
wb_data_df.to_csv("wb_data_df.csv")
print(len(wb_data_df))
wb_data_df.head(500)


```
